```python
#Periodo com maiores vvoumes de vendas por mÃªs
import pandas as pd
import plotly.express as px
from matplotlib import pyplot as plt

df_orders_dataset = pd.read_csv('assets/olist_orders_dataset.csv', encoding='latin1')
df_payments = pd.read_csv('assets/olist_order_payments_dataset.csv', encoding='latin1')
df_orders_dataset = df_orders_dataset.dropna()
df_payments = df_payments.dropna()

#Renomeando colunas da tabela olist_orders_dataset
df_orders_dataset = df_orders_dataset.rename(columns={
   'order_id': 'ID_do_pedido',
   'customer_id': 'ID_cliente',
   'order_status': 'Status_pedido',
    'order_purchase_timestamp': 'Data_da_compra',
    'order_approved_at': 'Pedido_aprovado_em',
    'order_delivered_carrier_date': 'Pedido_entregue_data_transportadora',
   'order_delivered_customer_date': 'Pedido_entregue_ao_cliente',
   'order_estimated_delivery_date': 'Data_estimada_de_entrega'

})
#Renomeando colunas da tabela olist_order_payments_dataset
df_payments = df_payments.rename(columns={
   'order_id': 'ID_pedido',
   'payment_sequential': 'Pagamento_sequencial',
   'payment_type': 'Forma_de_pagamento',
    'payment_installments': 'Parcelamento',
    'payment_value': 'Valor'
})

# Converter a coluna 'Data_da_compra' para datetime
df_orders_dataset['Data_da_compra'] = pd.to_datetime(df_orders_dataset['Data_da_compra'])

# Criar uma coluna 'AnoMes' para agrupar por mÃªs e ano
df_orders_dataset['AnoMes'] = df_orders_dataset['Data_da_compra'].dt.to_period('M')

# OrdenaÃ§Ã£o os dados por AnoMes
df_orders_dataset = df_orders_dataset.sort_values(by='AnoMes')
display(df_orders_dataset)

#Contagem de pedidos por mes
monthly_orders = df_orders_dataset.groupby('AnoMes').size().reset_index(name='Numero_de_Pedidos')
monthly_orders = monthly_orders.sort_values(by='AnoMes')
display(monthly_orders)

# Plotar o histograma
plt.figure(figsize=(12, 6))
ax = monthly_orders.plot(kind='bar', x='AnoMes', y='Numero_de_Pedidos', legend=False, ax=plt.gca())
ax.set(title='NÃºmero de Pedidos por MÃªs', xlabel='MÃªs/Ano', ylabel='NÃºmero de Pedidos')
ax.set_xticklabels(monthly_orders['AnoMes'], rotation=90)
plt.tight_layout()
plt.show()
```
